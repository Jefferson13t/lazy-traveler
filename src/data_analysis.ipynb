```python
import pandas as pd
```

```python
df = pd.read_csv('../data/feriados_com_pos.csv')
df['data'] = pd.to_datetime(df['data'])
print(df.shape)
df.head()
```

```python
df['cidade_uf'].unique(), len(df['cidade_uf'].unique())
```

```python
df['feriado'].unique(), len(df['feriado'].unique())
```

```python
df["mes"] = df["data"].dt.month
holidays_per_month = df["mes"].value_counts().sort_index()
holidays_per_month.plot(kind="bar", title="Número de feriados por mês", xlabel="Mês", ylabel="Número de feriados")
```
